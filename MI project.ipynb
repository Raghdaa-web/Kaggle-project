```python
sales= pd.get_dummies(sales, columns=['state_holiday'], prefix='state_holiday')
```

```python
sales.head()
```

Feateaure Engineering:

```python
import pandas as pd

# Convert the 'date' column to datetime format
sales['date'] = pd.to_datetime(sales['date'])

# Extract day, month, and year features
sales['day'] = sales['date'].dt.day
sales['month'] = sales['date'].dt.month
sales['year'] = sales['date'].dt.year

sales.head()
```

```python
sales = sales.drop('date', axis=1)
```

```python
sales.head()
```

```python
sales.dtypes
```

```python
sales_df= sales
```

```python
sales_df.rename(columns = {'nb_customers_on_day':'number_of_customers'}, inplace = True) 
sales_df.head()

#sales_df[nb_customers_on_day] = sales_df[number_of_customers]
```

**Model training**
#define the feature (x) and the target veriable (y)

```python
feature =['promotion', 'month', 'state_holiday_0', 'state_holiday_a',  'state_holiday_b', 'state_holiday_c', 'number_of_customers', 'day_of_week']
target = 'sales'
```

```python
x = sales_df[feature]
y = sales_df[target]
```

```python
from sklearn.model_selection import train_test_split #random state can be any number
X_train, X_val,y_train, y_val = train_test_split(x,y, test_size = 0.5, random_state=0)
```

```python
model = RandomForestRegressor (random_state=0)
```

Train the model

```python
model.fit(X_train, y_train)
```

```python
print(X_train.shape)
print(X_val.shape)
```

```python
predictions = model.predict(X_train)
```

**Evaluate the model: **

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(y_val, predictions)
mae = mean_absolute_error(y_val, predictions)

print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
```
